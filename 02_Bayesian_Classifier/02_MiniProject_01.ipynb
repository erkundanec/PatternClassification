## Question 01: Classify the data sample if mean and variance of the Gaussian distributions are known?

### Description
Given likelihood  

$p([x_1, x_2]^t |\omega_1) ∼ N([0,0]^t,4I), \\
p([x_1, x_2]^t |\omega_2) ∼ N([10,0]^t,4I), \\
p([x_1, x_2]^t |\omega_3) ∼ N([5,5]^t,5I),$

where $I$ is the $2 \times 2$ identity matrix. What is the error rate on the test set when the Bayesian decision rule is employed for classification?

### Bayes' Rule:
$ P(\omega_j|{\rm x}) = \frac{p({\rm x}|\omega_j) * P(\omega_j)}{p({\rm x})}$ 

with **prior probabilities**: $P(\omega_1) \; = P(\omega_2) \; = P(\omega_3)\; = \frac{1}{3}$
